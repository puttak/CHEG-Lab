The velocity profile of turbulent flows can be described by the 1/7 Power Law http://web.mit.edu/13.021/demos/lectures/lecture17.pdf:
$$\frac{V(r)}{V_{max}} = \bigg(1-\frac{r}{R}\bigg)^{1/7}$$

This profile was then integrated from 0 to R to give the average velocity. Divide this by the cross-sectional area to get
the total flow rate (Q) and rearrange to solve for $V_{max}$:
$$V_{max} = \frac{60}{49} \frac{Q}{R^2}$$
Plug in appropriate experimental values for Q and R to get $V_{max}$ and substitute back into velocity profile.

```python
% matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
# for small pipe
R = 0.0068326/2 # m

# laminar flow rate
qL = 300 # cm^3/min
QL = qL/100**3/60 # m^3/min

# turbulent flow rates
qT = np.array([800, 1000, 1500]) # cm^3/min
QT = qT/100**3/60 # m^3/min


# laminar maximum velocity
VmaxL = 2*QL/np.pi/R**2

# Turbulent maximum velocity
VmaxT = 60/49*QT/R**2 #4.138301885 # m/s

r = np.linspace(0,R,100)
```

```python
# Laminar velocity profile
VL = VmaxL*(1-(r/R)**2)
VL = np.append([VL[::-1]],[VL])

# Turbulent velocity profile
VT = []

for i in range(3):
    x = VmaxT[i]*(1-r/R)**(1/7)
    #print(x)
    VT.append(x)
    
for j in range(3):
    VT[j] = np.append([VT[j][::-1]],[VT[j]])
```

```python
r = np.append([-r[::-1]],[r])

plt.plot(VL,r,VT[0],r,VT[1],r,VT[2],r)
plt.title('Laminar and Turbulent Velocity Profiles\n for Small Pipe')
plt.xlabel('Vmax (m/s)')
plt.ylabel('Radius (m)')
plt.legend(['Q = 300','Q = 800','Q = 1000','Q = 1500'])
```
